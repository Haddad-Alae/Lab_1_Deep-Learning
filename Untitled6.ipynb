```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split


# Load your dataset
df = pd.read_csv('/content/predictive_maintenance.csv')

# Step 1: Impute or drop NaN values

# Identify numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Impute NaN values for numeric columns using median
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Impute NaN values for categorical columns using the mode (most frequent value)
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Step 2: Convert categorical columns to numeric (Label Encoding)
label_encoder = LabelEncoder()

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
# Step 3: Exploratory Data Analysis (EDA) - Visualize correlation matrix

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 4: Apply SMOTE for balancing the dataset (if the classes are imbalanced)
X = df.drop(columns=['Target', 'Failure Type'])  # Assuming 'Target' is the target column
y = df['Failure Type']  # Assuming 'Failure Type' is the class label column
```

```python
# Apply SMOTE for balancing the dataset
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Optional: Check the shape after resampling
print(f"Original data shape: {X.shape}, Resampled data shape: {X_resampled.shape}")

# Step 5: Split the resampled dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Step 6: Standardize the features (important for neural networks)
scaler = StandardScaler()

# Fit the scaler on training data and transform both train and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 7: Define the Deep Neural Network (DNN) model using PyTorch
import torch
import torch.nn as nn
import torch.optim as optim

class DNNModel(nn.Module):
    def __init__(self, input_size, output_size):
        super(DNNModel, self).__init__()
        self.layer1 = nn.Linear(input_size, 64)
        self.layer2 = nn.Linear(64, 32)
        self.layer3 = nn.Linear(32, output_size)
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.relu(self.layer2(x))
        x = self.layer3(x)
        return self.softmax(x)

# Initialize the model, loss function, and optimizer
input_size = X_train_scaled.shape[1]  # Number of features
output_size = len(y.unique())  # Number of classes
model = DNNModel(input_size, output_size)

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()  # For multi-class classification
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Step 8: Train the model

num_epochs = 20  # Set number of epochs
train_loss = []
train_accuracy = []
test_loss = []
test_accuracy = []

for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()

    # Convert data to torch tensors
    X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)

    # Forward pass
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)

    # Backward pass and optimization
    loss.backward()
    optimizer.step()

    # Compute accuracy
    _, predicted = torch.max(output, 1)
    correct = (predicted == y_train_tensor).sum().item()
    accuracy = correct / y_train_tensor.size(0)

    # Append training statistics
    train_loss.append(loss.item())
    train_accuracy.append(accuracy)

    # Test the model on the test set
    model.eval()
    with torch.no_grad():
        X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
        y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)

        test_output = model(X_test_tensor)
        test_loss_value = criterion(test_output, y_test_tensor)
        _, test_predicted = torch.max(test_output, 1)
        test_correct = (test_predicted == y_test_tensor).sum().item()
        test_accuracy_value = test_correct / y_test_tensor.size(0)

        test_loss.append(test_loss_value.item())
        test_accuracy.append(test_accuracy_value)

    # Print statistics every few epochs
    if (epoch + 1) % 5 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {accuracy:.4f}")
```

```python

# Step 9: Visualize the loss and accuracy curves
plt.figure(figsize=(12, 5))

# Plot Loss vs Epochs
plt.subplot(1, 2, 1)
plt.plot(range(num_epochs), train_loss, label="Train Loss")
plt.plot(range(num_epochs), test_loss, label="Test Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.legend()

# Plot Accuracy vs Epochs
plt.subplot(1, 2, 2)
plt.plot(range(num_epochs), train_accuracy, label="Train Accuracy")
plt.plot(range(num_epochs), test_accuracy, label="Test Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Epochs")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 10: Calculate metrics on both training and test datasets
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Training metrics
train_preds = model(torch.tensor(X_train_scaled, dtype=torch.float32))
_, train_preds = torch.max(train_preds, 1)
train_accuracy_score = accuracy_score(y_train, train_preds.numpy())
train_f1_score = f1_score(y_train, train_preds.numpy(), average='weighted')

# Test metrics
test_preds = model(torch.tensor(X_test_scaled, dtype=torch.float32))
_, test_preds = torch.max(test_preds, 1)
test_accuracy_score = accuracy_score(y_test, test_preds.numpy())
test_f1_score = f1_score(y_test, test_preds.numpy(), average='weighted')

print("Training Metrics:")
print(f"Accuracy: {train_accuracy_score:.4f}")
print(f"F1 Score: {train_f1_score:.4f}")

print("Test Metrics:")
print(f"Accuracy: {test_accuracy_score:.4f}")
print(f"F1 Score: {test_f1_score:.4f}")
```

```python
import torch.nn.functional as F

# Step 11: Modify the model to include L2 Regularization (Weight Decay) and Dropout

class DNNModelWithRegularization(nn.Module):
    def __init__(self, input_size, output_size, dropout_rate=0.5, weight_decay=1e-4):
        super(DNNModelWithRegularization, self).__init__()
        self.layer1 = nn.Linear(input_size, 64)
        self.layer2 = nn.Linear(64, 32)
        self.layer3 = nn.Linear(32, output_size)

        # Dropout layer
        self.dropout = nn.Dropout(p=dropout_rate)

        # Hyperparameter for weight decay (L2 regularization)
        self.weight_decay = weight_decay
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.dropout(x)  # Apply dropout after the first layer
        x = self.relu(self.layer2(x))
        x = self.dropout(x)  # Apply dropout after the second layer
        x = self.layer3(x)
        return self.softmax(x)

# Initialize the model with dropout and L2 regularization
model_with_reg = DNNModelWithRegularization(input_size, output_size, dropout_rate=0.5, weight_decay=1e-4)

# Loss function and optimizer with weight decay (L2 regularization)
criterion = nn.CrossEntropyLoss()  # For multi-class classification
optimizer = optim.Adam(model_with_reg.parameters(), lr=0.001, weight_decay=1e-4)  # weight_decay is the L2 regularization strength

# Step 12: Train the model with regularization techniques

num_epochs = 20  # Set number of epochs
train_loss = []
train_accuracy = []
test_loss = []
test_accuracy = []

for epoch in range(num_epochs):
    model_with_reg.train()
    optimizer.zero_grad()

    # Convert data to torch tensors
    X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)

    # Forward pass
    output = model_with_reg(X_train_tensor)
    loss = criterion(output, y_train_tensor)

    # Backward pass and optimization
    loss.backward()
    optimizer.step()

    # Compute accuracy
    _, predicted = torch.max(output, 1)
    correct = (predicted == y_train_tensor).sum().item()
    accuracy = correct / y_train_tensor.size(0)

    # Append training statistics
    train_loss.append(loss.item())
    train_accuracy.append(accuracy)

    # Test the model on the test set
    model_with_reg.eval()
    with torch.no_grad():
        X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
        y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)

        test_output = model_with_reg(X_test_tensor)
        test_loss_value = criterion(test_output, y_test_tensor)
        _, test_predicted = torch.max(test_output, 1)
        test_correct = (test_predicted == y_test_tensor).sum().item()
        test_accuracy_value = test_correct / y_test_tensor.size(0)

        test_loss.append(test_loss_value.item())
        test_accuracy.append(test_accuracy_value)

    # Print statistics every few epochs
    if (epoch + 1) % 5 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {accuracy:.4f}")

# Step 13: Visualize the loss and accuracy curves
plt.figure(figsize=(12, 5))

# Plot Loss vs Epochs
plt.subplot(1, 2, 1)
plt.plot(range(num_epochs), train_loss, label="Train Loss")
plt.plot(range(num_epochs), test_loss, label="Test Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.legend()

# Plot Accuracy vs Epochs
plt.subplot(1, 2, 2)
plt.plot(range(num_epochs), train_accuracy, label="Train Accuracy")
plt.plot(range(num_epochs), test_accuracy, label="Test Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Epochs")
plt.legend()

plt.tight_layout()
plt.show()

# Step 14: Evaluate model performance using accuracy and F1-score
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Training metrics
train_preds = model_with_reg(torch.tensor(X_train_scaled, dtype=torch.float32))
_, train_preds = torch.max(train_preds, 1)
train_accuracy_score = accuracy_score(y_train, train_preds.numpy())
train_f1_score = f1_score(y_train, train_preds.numpy(), average='weighted')

# Test metrics
test_preds = model_with_reg(torch.tensor(X_test_scaled, dtype=torch.float32))
_, test_preds = torch.max(test_preds, 1)
test_accuracy_score = accuracy_score(y_test, test_preds.numpy())
test_f1_score = f1_score(y_test, test_preds.numpy(), average='weighted')

print("Training Metrics with Regularization:")
print(f"Accuracy: {train_accuracy_score:.4f}")
print(f"F1 Score: {train_f1_score:.4f}")

print("Test Metrics with Regularization:")
print(f"Accuracy: {test_accuracy_score:.4f}")
print(f"F1 Score: {test_f1_score:.4f}")
```
