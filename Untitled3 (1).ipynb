```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/prices.csv')
```

```python
# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# Check for missing values
print(df.isnull().sum())

# Summary statistics
print(df.describe())
```

```python
# Visualize data
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='close', data=df)
plt.title('Stock Close Prices Over Time')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split

# Prepare the data
df = df.dropna(subset=['close', 'open', 'high', 'low', 'volume'])  # Drop rows with NaN
features = df[['open', 'high', 'low', 'volume']].values  # Feature columns
target = df['close'].values  # Target column

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# Define the model
class DNN(nn.Module):
    def __init__(self):
        super(DNN, self).__init__()
        self.layer1 = nn.Linear(4, 64)
        self.layer2 = nn.Linear(64, 32)
        self.layer3 = nn.Linear(32, 1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.relu(self.layer2(x))
        x = self.layer3(x)
        return x

# Instantiate the model
model = DNN()
```

```python
# Define the number of epochs
epochs = 100

# Initialize the optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)  # Example: Adam optimizer with learning rate 0.001
criterion = nn.MSELoss()

# Track loss and accuracy during training
train_losses = []
test_losses = []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)
    loss.backward()
    optimizer.step()

    train_losses.append(loss.item())

    # Evaluate on the test set
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test_tensor)
        test_loss = criterion(y_pred, y_test_tensor)
        test_losses.append(test_loss.item())

# Plot Loss vs Epochs
plt.figure(figsize=(10, 6))
plt.plot(range(epochs), train_losses, label='Training Loss')
plt.plot(range(epochs), test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs')
plt.legend()
plt.show()
```

```python
class DNN_with_Dropout(nn.Module):
    def __init__(self):
        super(DNN_with_Dropout, self).__init__()
        self.layer1 = nn.Linear(4, 64)
        self.dropout = nn.Dropout(0.5)  # Dropout layer
        self.layer2 = nn.Linear(64, 32)
        self.layer3 = nn.Linear(32, 1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = self.dropout(x)  # Apply dropout after first layer
        x = torch.relu(self.layer2(x))
        x = self.layer3(x)
        return x
```

```python
# Initialize the original model
model = DNN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
train_losses_original = []
test_losses_original = []

# Train the original model (DNN)
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)
    loss.backward()
    optimizer.step()

    train_losses_original.append(loss.item())

    # Evaluate on the test set
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test_tensor)
        test_loss = criterion(y_pred, y_test_tensor)
        test_losses_original.append(test_loss.item())

# Initialize the model with dropout
model_with_dropout = DNN_with_Dropout()
optimizer_dropout = torch.optim.Adam(model_with_dropout.parameters(), lr=0.001)
train_losses_dropout = []
test_losses_dropout = []

# Train the model with dropout
for epoch in range(epochs):
    model_with_dropout.train()
    optimizer_dropout.zero_grad()
    output = model_with_dropout(X_train_tensor)
    loss = criterion(output, y_train_tensor)
    loss.backward()
    optimizer_dropout.step()

    train_losses_dropout.append(loss.item())

    # Evaluate on the test set
    model_with_dropout.eval()
    with torch.no_grad():
        y_pred = model_with_dropout(X_test_tensor)
        test_loss = criterion(y_pred, y_test_tensor)
        test_losses_dropout.append(test_loss.item())

# Plot Loss vs Epochs for both models
plt.figure(figsize=(12, 6))

# Plot for original model (DNN)
plt.subplot(1, 2, 1)
plt.plot(range(epochs), train_losses_original, label='Training Loss (Original)')
plt.plot(range(epochs), test_losses_original, label='Test Loss (Original)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs (Original Model)')
plt.legend()

# Plot for model with dropout (DNN_with_Dropout)
plt.subplot(1, 2, 2)
plt.plot(range(epochs), train_losses_dropout, label='Training Loss (With Dropout)')
plt.plot(range(epochs), test_losses_dropout, label='Test Loss (With Dropout)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs (With Dropout)')
plt.legend()

plt.tight_layout()
plt.show()

# Optional: Print final test losses for comparison
print(f"Final Test Loss (Original Model): {test_losses_original[-1]}")
print(f"Final Test Loss (With Dropout): {test_losses_dropout[-1]}")
```
